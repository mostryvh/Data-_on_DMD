üß¨ Medical Data Analysis for **Duchenne Muscular Dystrophy (DMD)**

## üß† Function: `create_Age_at_diagnosis(row)`

### üéØ Purpose:

To generate a **logical age of diagnosis** that corresponds to the
**current age** of the patient.

### üß© Medical Logic:

-   DMD is usually **diagnosed between 3 and 10 years old**.\
-   The **diagnosis age** cannot exceed the **current age**.\
-   Minimum possible diagnosis age: **3 years** (rarely earlier).\
-   Maximum possible diagnosis age: **10 years** (often detected
earlier).

``` python
def create_Age_at_diagnosis(row):
"""
Function to generate the age of diagnosis based on the current age
using logical constraints consistent with DMD clinical patterns.
"""
current_age = row['Age']
max_age_rand = min(10, current_age)
if max_age_rand >= 3:
return np.random.randint(3, max_age_rand + 1)
else:
return 3
```

## ‚öß Gender Distribution

The percentage of **males** is intentionally higher than **females**,
because: - **Males** are the ones who **develop and exhibit** the
disease.\
- **Females** are typically **carriers** of the defective gene and
usually **do not show symptoms**.

## üß´ Gene Mutation Types

Three main mutation types are seen in DMD, each with a different
frequency and severity:

Mutation Type        Incidence Rate   Severity Level
-------------------- ---------------- --------------------
**Deletion**         65--70%          Moderate to Severe
**Point Mutation**   20--25%          Variable
**Duplication**      10--15%          Mild to Moderate

## üë®üë©üë¶ Family History

-   Approximately **30--40%** of DMD cases are **inherited** (positive
family history).\
-   The remaining **60--70%** are due to **new mutations (de novo)**
with **no family history**.

------------------------------------------------------------------------

## üß™ Creatine Kinase (CK) Levels

-   In children with DMD, **CK levels** peak at **10--20√ó the upper
normal limit** around the age of **2 years**.\
-   Levels then **decrease by about 25% per year** as muscle tissue is
gradually replaced by **fat and fibrotic tissue**.\
-   In **female carriers**, CK levels are **normal**.\
-   In **advanced disease**, CK levels may **return to normal** due to
extensive muscle loss.

------------------------------------------------------------------------

## üíä Steroid Therapy

### üìä Treatment Statistics:

-   In a large U.S. study, **84% of males with DMD** received steroid
therapy by **age 8** (peak usage).\
-   The rate of steroid use **gradually declines** after **age 10**.

### üö´ Reasons for Not Starting Therapy:

-   Parental concerns about **side effects**.\
-   **Lack of a medical prescription**.

------------------------------------------------------------------------

### üïí Optimal Time to Begin Treatment:

-   **Age Range:** Typically between **4--6 years old**.\
-   **Functional Stage:** When **motor skills** (running, jumping) have
**plateaued or begun to decline**, but **independent walking is
still possible**.\
-   This corresponds to the **intermediate stage** of disease
progression.

------------------------------------------------------------------------
## ‚ù§ Left Ventricular Ejection Fraction (LVEF%)

### üîª Patients **Not Receiving Steroid Therapy**

-   **Faster deterioration** of cardiac function.\
-   **Heart failure** develops **earlier and more frequently**.\
-   **Onset:** Usually between **10--14 years old**.\
-   By **age 18**, almost all patients show **reduced LVEF**.\
-   **Values:**
-   LVEF \< 55% ‚Üí indicates dysfunction.\
-   LVEF \< 40% or \< 30% ‚Üí indicates severe heart failure.

üìå **Summary:**\
Without steroids, cardiac deterioration is **rapid and progressive**,
leading to **early heart failure** and a **sharp LVEF decline**.

------------------------------------------------------------------------

### üí™ Patients **Receiving Steroid Therapy**

-   **Clear protective effect:** Steroids (e.g., *prednisone*,
*deflazacort*) slow skeletal muscle atrophy and protect **cardiac
muscle**.\
-   **Slower LVEF decline:** Steroid therapy **delays the onset** of
cardiomyopathy for several years.\
-   **Lower incidence of heart failure.**\
-   **LVEF remains \>50--55%** for a longer time.\
-   Even when decline begins, it occurs **gradually**, reducing risk of
severe cardiac failure.

üìå **Summary:**\
Steroid therapy **significantly reduces** the rate and severity of
**cardiomyopathy in DMD**, delays its onset, and **preserves LVEF**
within healthy ranges for longer periods.

------------------------------------------------------------------------

## üåü Overall Summary

-   DMD is a **genetically complex and progressive** disorder.\
-   **Early diagnosis** and **appropriate steroid therapy** dramatically
**improve life quality and outcomes**.\
-   **Data-driven analysis** of clinical variables (age, mobility, gene
type, therapy) allows for **accurate modeling** of disease
progression.

````python
#1. Determine the number of rows
num_rows = 600000

#2. Create Patient Identifier
patient_ids = [f'DMD_{i:06}' for i in range(1, num_rows+1)] # use DMD_{i:06} to the id start with DMD and 06 to input six 0

#3. Create a lifetime column (from 5 to 29 years)
Age_column = np.random.randint(5, 30, size=num_rows)

#4. Creation DataFrame
dmd = pd.DataFrame({
    'Patient_ID': patient_ids,
    'Age': Age_column
})
````

````python
#5. Create a column Age_at_diagnosis
def create_Age_at_diagnosis(row):
    current_age = row['Age']
    max_age_rand = min(10, current_age)

    if max_age_rand >= 3:
        return np.random.randint(3, max_age_rand + 1)
    else:
        return 3

dmd['Age_at_diagnosis'] = dmd.apply(create_Age_at_diagnosis, axis=1)
````

````python
#6. create Duration_of_DMD column
dmd['Duration_of_DMD'] = dmd['Age'] - dmd['Age_at_diagnosis']


#7. Create a sex column

dmd['Sex'] = np.random.choice(
    ['Male', 'Female'],
    size=len(dmd),
    p=[0.85, 0.15]
)
````

````python
#8. Create motion status column
dmd['Ambulatory_Status'] = None

#Step 1: Children < 10 years ‚Üí are able to walk
#.loc is a tool for accessing specific rows and columns in DataFrame

dmd.loc[dmd['Age'] < 10, 'Ambulatory_Status'] = 'Ambulatory'

#Step 2: Adults >15 years ‚Üí Lost power
dmd.loc[dmd['Age'] > 15, 'Ambulatory_Status'] = 'Lost_Ambulation'

 #Step 3: Transition (10-12 years)
Transition_1 = (dmd['Age'] >= 10) & (dmd['Age'] <= 12)
dmd.loc[Transition_1, 'Ambulatory_Status'] = np.random.choice(
    ['Ambulatory', 'Lost_Ambulation'],
    size=Transition_1.sum(),
    p=[0.75, 0.25]
)
#Step 4: Transition (13-15 years)
Transition_2 = (dmd['Age'] >= 13) & (dmd['Age'] <= 15)
dmd.loc[Transition_2, 'Ambulatory_Status'] = np.random.choice(
    ['Ambulatory', 'Lost_Ambulation'],
    size=Transition_2.sum(),
    p=[0.20, 0.80]
)


#Step 5: Females ‚Üí Always able to walk
dmd.loc[dmd['Sex'] == 'Female', 'Ambulatory_Status'] = 'Ambulatory'
````

````python
#9. create Gene_Mutation column
dmd['Gene_Mutation']=np.random.choice(
    ['Deletion','Duplication','Point Mutation'],
    size=len(dmd),
    p=[0.68, 0.12, 0.20]  # 68% Deletion, 12% Duplication, 20% Point
    )
````

````python
#10. create Exon_Mutation column
def create_mutation_exon(row):
    mutation_type = row['Gene_Mutation']
    if mutation_type == 'Deletion':
        if np.random.random() < 0.7:
            start = np.random.choice([2, 3, 45, 48, 51])
            length = np.random.randint(3, 8)
            return f'Exon {start}-{start+length}'
        else:
            exon = np.random.choice([45, 46, 47, 48, 49, 50, 51, 52, 53])
            return f'Exon {exon}'
    elif mutation_type == 'Duplication':
        exon = np.random.choice([2, 8, 44])
        return f'Exon {exon}'
    else:
        exon = np.random.randint(1, 80)
        return f'Exon {exon}'

dmd['Mutation_Exon'] = dmd.apply(create_mutation_exon, axis=1)
````

````python
#11. create Family_History column
 # we make persentg 35%Present
dmd['Family_History']=np.random.choice(
     ['Present', 'Absent'],
     size=len(dmd),
     p=[0.35, 0.65])
````

````python
#12. create ck_level column
def create_ck_level(row):

#    CK Level depends on:
#   - Sex (Females: normal range)
#   - Age (younger patients: higher CK)
#   - Ambulatory_Status (ambulatory: higher CK)
#   - Duration (longer duration: lower CK)

    if row['Sex'] == 'Female':
        return np.random.randint(100, 300)

    # Males with DMD
    age = row['Age']
    ambulatory = row['Ambulatory_Status']
    duration = row['Duration_of_DMD']

    # Young children (5-10 years): Very high CK
    if age <= 10:
        return np.random.randint(15000, 25000)

    # Teenagers (11-15 years): High CK
    elif age <= 15:
        if ambulatory == 'Ambulatory':
            return np.random.randint(10000, 18000)
        else:  # Lost_Ambulation
            return np.random.randint(7000, 12000)

    # Adults (16+ years): Lower CK
    else:
        if duration <= 10:
            return np.random.randint(5000, 10000)
        else:  # duration > 10
            return np.random.randint(3000, 7000)

dmd['CK_Level'] = dmd.apply(create_ck_level, axis=1)
````

````python
#13. create higth column
"""
 Depending on the age column, we'll divide it into groups of tired and old children.
 male and famele
 avg higth to avrey group {1 famele[
  5->10 year Children higth 105->145
  11->18 Teenagers higth 140->170
  18->30 Adults higth 155->175]

 2 male[
   5->10 year Children higth 100->130
   11->18 Teenagers higth 125->155
   18->30 Adults higth   140->170       ]}
 this column depends on columns of age& sex& ambulatory duration
"""
def create_higth(row):

#  Height depends on Age and Sex.
# make a column in varubile

    age = row['Age']
    sex = row['Sex']

    if sex == 'Female':
        if age >= 5 and age <= 10:
            return np.random.randint(105, 145)
        elif age >= 11 and age <= 18:
            return np.random.randint(140, 170)
        else:  # age > 18
            return np.random.randint(155, 175)
    else:  # Male with DMD
        if age >= 5 and age <= 10:
            return np.random.randint(100, 130)
        elif age >= 11 and age <= 18:
            return np.random.randint(125, 155)
        else:  # age > 18
            return np.random.randint(140, 168)

dmd['Height_cm'] = dmd.apply(create_higth, axis=1)
````

````python
#14. create Weight column
"""
 we come do the same thing
 Depending on the age column, we'll divide n.
 it into groups of tired and old childre male and famele
 avg weight to avrey group
  1 famele [
  5->10 year Children weight 16->40
  11->18 Teenagers weight 35->65
  18->30 Adults weight  ]  50->75

 2 male[
   5->10 year Children weight 16->32
   11->18 Teenagers weight 30->52
   18->30 Adults weight   40->65
   if Lost_Ambulation to age 11->18 the wight 25->45
   if Lost_Ambulation to age 18->30  depends of column duration
   if duration <=10 this recently Lost_Ambulation weigth 35->55
   if duration >10 weigth 30->50 ]}
 """
def create_weight(row):
    # this column depends on column Age& sex& ambulatory % duration

    age = row['Age']
    sex = row['Sex']
    ambulatory = row['Ambulatory_Status']
    duration = row['Duration_of_DMD']

    if sex == 'Female':  # Females without DMD
        if age >= 5 and age <= 10:
            return np.random.randint(16, 40)
        elif age >= 11 and age <= 18:
            return np.random.randint(35, 65)
        else:  # age > 18
            return np.random.randint(50, 75)

    else:  # Male with DMD
        if age >= 5 and age <= 10:  # Before Lost_Ambulation
            return np.random.randint(16, 32)

        elif age >= 11 and age <= 18:  # Beginning of Lost_Ambulation
            if ambulatory == 'Ambulatory':
                return np.random.randint(30, 52)
            else:  # Lost_Ambulation
                return np.random.randint(25, 45)

        else:  # age > 18
            if ambulatory == 'Ambulatory':
                return np.random.randint(40, 65)
            else:  # Lost_Ambulation depends on duration
                if duration <= 10:  # Recently Lost_Ambulation
                    return np.random.randint(35, 55)
                else:  # Old Lost_Ambulation (duration > 10)
                    return np.random.randint(30, 50)

dmd['Weight_kg'] = dmd.apply(create_weight, axis=1)
````

````python
#15. create colum of steroid_therapy
"""
in column value it yes or no
Logic:

- Most patients (70-80%) receive steroids

- Young children: Higher percentage

- Those with paralysis: May continue or stop

- Females: Do not require treatment
- this column It has an effect on FVC and Cardiac_Function'
Steroid therapy based on American study data:
    - Peak usage (84%) at age 7-9
    - Gradual decline after age 10
    - Females: never receive treatment (carriers only)
    We will take the following columns into consideration.
    Sex & Age & Ambulatory_Status
    make it in varubile
"""

def create_steroid(row):

    age = row['Age']
    sex = row['Sex']
    ambulatory = row['Ambulatory_Status']
    # the female do not receive treatment
    if sex == 'Female':
      return 'NO'
      # male Males aged 5-6 often require treatment for AIDS.
    else :
      if age >=5 and age <=6: # beginning of Treatment
        return np.random.choice(['Yes', 'NO'],p=(0.60,0.40))
      elif age >=7 and age <=9: #Highest stage of treatment
        return np.random.choice(['Yes', 'NO'],p=(0.84,0.16))
      elif age >=10 and age <=12: # Gradual decline after age 10
        return np.random.choice(['Yes', 'NO'],p=(0.75,0.25))
      elif age >=13 and age <=18: #  Continued deterioration
        return np.random.choice(['Yes', 'NO'],p=(0.60,0.40))
      else: #For cases older than 18, we examine the movement status.
        if ambulatory == 'Ambulatory': # in this case the treatment
          return np.random.choice(['Yes','NO'],p=(0.50,0.50))


dmd['Steroid_therapy'] = dmd.apply(create_steroid, axis=1)
````

````python
#16. create column LVEF%%
"""
 This column is based on the Steroid_therapy column and the age column.
  If Steroid_therapy is marked "no," the heart condition is serious, with a
  rate ranging from 25% to 40%. If it is marked "yes," the severity depends on
  age. Children aged 5 to 10 years have good heart health, with a rate of 80%
  or higher. From 10 to 15 years old, their heart condition tends to be fair,
   with the onset of symptoms, resulting in a rate of 55%. Individuals older
   than 15 years will have the highest rate of heart disease,
   ranging from 25% to 40%.

 """
def create_lvef(row):
  age = row['Age']
  steroid = row['Steroid_therapy']


  if steroid == 'No':
  #Faster in the task job

    if age >=5 and age <=10:
      return np.random.randint(60,70)
    elif age >=11 and age <=15:
      return np.random.randint(40,50)
    else:
      return np.random.randint(25,45)
  else:
    if age >=5 and age <=10: #have good heart health
      return np.random.randint(80,100)
    elif age >=11 and age <=15: # Moderate, with the appearance of symptoms
      return np.random.randint(55,65)
    else:  #have the highest rate of heart disease
      return np.random.randint(35,50)

dmd['LVEF_%']= dmd.apply(create_lvef,axis=1)
````

````python
#17.create coulmn Cardiomyopathy
"""
this column It contains three values
Absent
moderate
Severe
It depends on create_lvef
"""
def create_Cardiomyopathy(row):
  lvef = row['LVEF_%']
  if lvef >=25 and lvef <=40:
    return 'Severe'
  elif lvef >=55 and lvef <=65:
    return 'Moderate'
  elif lvef>=80 and lvef <=100:
    return 'Absent'

dmd['Cardiomyopathy'] = dmd.apply(create_Cardiomyopathy,axis=1)
````

````python
#18. create FVC column
"""
This column is based on the Steroid_therapy column and the Age column.
If Steroid_therapy is marked "no," lung function declines more rapidly,
with FVC values ranging from 30% to 50%.

If Steroid_therapy is marked "yes," the level of lung function depends on age:

Children aged 5 to 10 years generally have good pulmonary function, with FVC
values ranging from 80% to 100%.

Individuals aged 11 to 15 years show a moderate decline, with FVC values ranging
 from 60% to 80%, reflecting the beginning of respiratory muscle weakness.

Those older than 15 years tend to have a marked reduction in lung capacity, with
 FVC values between 40% and 60%.
"""

def create_FVC(row):

  age = row['Age']
  steriod = row['Steroid_therapy']
  if steriod == 'No':
#Faster in the task job
        if age >=5 and age <=10:
            return np.random.randint(70, 90)
        elif age >=11 and age <=15:
            return np.random.randint(50, 70)
        else:
            return np.random.randint(30, 50)
  else:
# Cortisone preserves lung function for a longer period
    if age >=5 and age <=10:
      return np.random.randint(80, 100)
    elif age >=11 and age <=15:
      return np.random.randint(60, 80)
    else:
      return np.random.randint(40, 60)


dmd['FVC_%'] = dmd.apply(create_FVC, axis=1)
````

````python
#19. create Respiratory column
def create_respiratory_status(row):
    fvc = row['FVC_%']
    if fvc >=30 and fvc  <= 50:
        return 'Severe'
    elif fvc >50 and fvc  <= 70:
        return 'Moderate'
    elif fvc > 70:
        return 'Normal'

dmd['Respiratory_status'] = dmd.apply(create_respiratory_status, axis=1)
````

````python
#20 create IQ_Score column
"""
Scientific Background:

Approximately 30‚Äì40% of children with DMD have learning difficulties
 or a slight intellectual disability.

In most cases, the IQ is between 80 and 90, and sometimes below 70
(mild intellectual disability).

This is because the dystrophin gene is affected in the brain as
 well as in the muscles.

The use of corticosteroids does not usually affect intelligence,
but age may indicate the extent of cognitive decline over time.
"""

def create_iq_score(row):
    age = row['Age']
    if age >= 5 and age <=10:
        return np.random.randint(85, 105)
    elif age >= 11 and age <=15 :
        return np.random.randint(75, 90)
    else:
        return np.random.randint(65, 80)

dmd['IQ_Score']= dmd.apply(create_iq_score,axis=1)
````

````python
#21. create Speech delay column
"""
Scientific Background:

Speech delay is very common in DMD, often appearing between the ages of 2 and 5.

Approximately 60% of children experience mild to moderate language delay.

The delay is often associated with lower IQ, not just age.
"""
def create_speech_delay(row):
    iq = row['IQ_Score']
    if iq < 70:
        return 'Severe'
    elif 70 <= iq < 85:
        return 'Moderate'
    else:
        return 'Normal'
dmd['Speech_delay'] = dmd.apply(create_speech_delay, axis=1)
````

````python
 #22. create outcome colum
"""
In DMD, "outcome" usually refers to the overall condition or prognosis.

It depends on:

Heart condition (LVEF%)

Lung condition (FVC%)

Corticosteroid use
"""

def create_outcome(row):
    lvef = row['LVEF_%']
    fvc = row['FVC_%']
    steroid = row['Steroid_therapy']

# Overall Health Score = Average Heart and Lung Health
    avg_function = (lvef + fvc) / 2

    if avg_function >= 70:
        return 'Good'
    elif 50 <= avg_function < 70:
        return 'Moderate'
    else:
#If cortisone is not used, the condition will be worse.
        if steroid == 'No':
            return 'Poor'
        else:
            return 'Moderate'
dmd['OutCome']=dmd.apply(create_outcome, axis=1)
````
